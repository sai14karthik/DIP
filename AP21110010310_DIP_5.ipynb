a. Develop programs for following intensity transformation operation on a grayscale image. Collect any gray scale image from any source. Process that image using these operations Image negative


```python
from google.colab.patches import cv2_imshow
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("/content/drive/MyDrive/DIP /Lenna_(test_image).png", cv2.IMREAD_GRAYSCALE)

cv2_imshow(img)
```

```python
def image_negative(img):
    L = 255
    neg_img = L - 1 - img

    return neg_img

neg_img = image_negative(img)

plt.imshow(neg_img, cmap='gray')
plt.title('Negative Image')
plt.show()
```

b. Log transformation and inverse log transform:

```python
def log_transform(img, c):
    img_log = c * np.log(1 + img)
    img_log = np.uint8(img_log)
    return img_log

def inverse_log_transform(img, c):
    img_inv_log = np.exp(img / c) - 1
    img_inv_log = np.uint8(img_inv_log)
    return img_inv_log

c = 1
log_img = log_transform(img, c)

inv_log_img = inverse_log_transform(log_img, c)

plt.imshow(log_img, cmap='gray')
plt.title('Log Transformation')
plt.show()

plt.imshow(inv_log_img, cmap='gray')
plt.title('Inverse Log Transformation')
plt.show()
```

c. Power law transformation:
Study the effect of different values of Gamma used in this transformation.

```python
def power_law_transform(img, gamma):
    img_float = img / 255.0
    img_gamma = np.power(img_float, gamma)
    img_gamma = np.uint8(img_gamma * 255)
    return img_gamma

gamma_1 = 0.5
gamma_2 = 2.0

power_img1 = power_law_transform(img, gamma_1)
power_img2 = power_law_transform(img, gamma_2)

plt.imshow(power_img1, cmap='gray')
plt.title('Power Law Transform (Gamma = 0.5)')
plt.show()

plt.imshow(power_img2, cmap='gray')
plt.title('Power Law Transform (Gamma = 2.0)')
plt.show()
```

d. Contrast stretching

```python
def contrast_stretching(img):
    min_val = np.min(img)
    max_val = np.max(img)

    stretched_img = (img - min_val) * (255 / (max_val - min_val))
    stretched_img = np.uint8(stretched_img)

    return stretched_img

contrast_img = contrast_stretching(img)

plt.imshow(contrast_img, cmap='gray')
plt.title('Contrast Stretching')
plt.show()
```

e. Gray level slicing

```python
def gray_level_slicing(img, low, high):
    sliced_img = np.zeros_like(img)
    sliced_img[(img >= low) & (img <= high)] = 255
    return sliced_img

low = 100
high = 200
sliced_img = gray_level_slicing(img, low, high)

plt.imshow(sliced_img, cmap='gray')
plt.title(f'Gray Level Slicing ({low}-{high})')
plt.show()
```
