a. Read, Display and write
any color image in other formats.

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```python
img_color = cv2.imread('/content/Lenna_(test_image).png', cv2.IMREAD_COLOR)
img_grayscale = cv2.imread('/content/Lenna_(test_image).png', cv2.IMREAD_GRAYSCALE)
img_unchanged = cv2.imread('/content/Lenna_(test_image).png', cv2.IMREAD_UNCHANGED)
```

```python
cv2_imshow(img_color)
cv2_imshow(img_grayscale)
cv2_imshow(img_unchanged)
```

Find RED, GREEN and BLUE plane of the
color image

```python
img = cv2.imread(r'/content/Lenna_(test_image).png')
```

```python
cv2_imshow(img)
```

```python
row, col, plane = img.shape
```

```python
temp = np.zeros((row, col, plane), np.uint8)
```

```python
temp[:, :, 0] = img[:, :, 0]
```

```python
cv2_imshow(temp)
```

```python
temp = np.zeros((row, col, plane), np.uint8)
```

```python
temp[:, :, 1] = img[:, :, 1]
```

```python
cv2_imshow(temp)
```

```python
temp = np.zeros((row, col, plane), np.uint8)
```

```python
temp[:, :, 2] = img[:, :, 2]
```

```python
cv2_imshow(temp)
```
