a. Averaging: Implement
averaging filtering operations for different window sizes and study their
effect on the quality of output image. Write your observations on output
image quality.


```python
from google.colab.patches import cv2_imshow
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def apply_averaging_filter(image, kernel_size):
    return cv2.blur(image, (kernel_size, kernel_size))

image_path = '/content/drive/MyDrive/DIP /Lenna_(test_image).png'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

kernel_sizes = [3, 5, 7, 9]

filtered_images = [apply_averaging_filter(image, k) for k in kernel_sizes]

plt.figure(figsize=(10, 10))

plt.subplot(3, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

for i, (filtered_img, k) in enumerate(zip(filtered_images, kernel_sizes)):
    plt.subplot(3, 3, i + 2)
    plt.imshow(filtered_img, cmap='gray')
    plt.title(f'Filtered Image (Kernel: {k}x{k})')
    plt.axis('off')

plt.tight_layout()
plt.show()
```


b. Weighted averaging:
Implement weighted averaging filtering operations for different window sizes
and study their effect on the quality of output image. Write your
observations on output image quality.


```python
def apply_custom_weighted_filter(image, kernel):
    return cv2.filter2D(image, -1, kernel)

custom_kernel = np.array([[1, 2, 1],
                          [2, 4, 2],
                          [1, 2, 1]], dtype=np.float32)
custom_kernel /= custom_kernel.sum()




image_path = '/content/drive/MyDrive/DIP /Lenna_(test_image).png'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
custom_filtered_image = apply_custom_weighted_filter(image, custom_kernel)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(custom_filtered_image, cmap='gray')
plt.title('Custom Weighted Filter (3x3)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

d. Max filtering

```python


def apply_max_filter(image, kernel_size):
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    return cv2.dilate(image, kernel)


image_path = '/content/drive/MyDrive/DIP /Lenna_(test_image).png'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

kernel_sizes = [3, 5, 7, 9]

filtered_images = [apply_max_filter(image, k) for k in kernel_sizes]

plt.figure(figsize=(10, 10))

plt.subplot(3, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

for i, (filtered_img, k) in enumerate(zip(filtered_images, kernel_sizes)):
    plt.subplot(3, 3, i + 2)
    plt.imshow(filtered_img, cmap='gray')
    plt.title(f'Max Filter (Kernel: {k}x{k})')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
